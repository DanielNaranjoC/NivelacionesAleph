### Curvas paramétricas

$(x(t), y(t))$

Dada la curva:

$x(t) = \sin(t)$

$y(t) = \cos(t)$

```python
t = np.linspace(0, 2*np.pi)
x = np.sin(t)
y = np.cos(t)
plt.plot(x,y)
plt.show()
```

$x(t)=\sin(t)\left(e^{\cos(t)}-2\cos(4t)+\sin\left(\frac{t}{12}\right)^5\right)$

$y(t)=\cos(t)\left(e^{\cos(t)}-2\cos(4t)+\sin\left(\frac{t}{12}\right)^5\right)$

```python
t = np.linspace(0, 2*np.pi, 200)
x = np.sin(t)*(np.exp(np.cos(t)) - 2*np.cos(4*t) + np.sin(t/12)**5)
y = np.cos(t)*(np.exp(np.cos(t)) - 2*np.cos(4*t) + np.sin(t/12)**5)
plt.plot(x,y)
plt.show()
```

#### Tangente

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.tan(x)

tol = 2*np.pi

y[y > tol] = np.nan
y[y < -tol] = np.nan

plt.plot(x,y)
plt.axis([-2*np.pi, 2*np.pi, -0.9*tol, 0.9*tol])
plt.show()
```

## Gráficas polares

$r = 3 \cos^2{(0.5\theta)}$ dónde $\theta\in[0, 2\pi]$

```python
from numpy import *
```

```python
theta = linspace(0, 2*pi)
r = 3*cos(0.5*theta)**2
plt.polar(theta, r, '-.r')
plt.show()
```

$r^2 = \sin(2\theta)$

```python
t = linspace(0, 2*pi, 1000)
r1 = sqrt(sin(2*t))
r2 = sqrt(-sin(2*t))
plt.polar(t,r1, 'r', t, r2, 'r')
plt.show()
```

### Curvas en 3D

\begin{cases}
z = z\\
x = r\cos(\theta)\\
y = r\sin(\theta)
\end{cases}

con $z\in[-2,2],\ \theta\in[-4\pi,4\pi]$ y $r=z^2+1$

```python
ax = plt.axes(projection='3d')
theta = linspace(-4*pi, 4*pi, 1000)
z = linspace(-2,2,1000)
r = z**2 + 1
x = r * cos(theta)
y = r* sin(theta)

ax.plot(x,y,z)

ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.set_zlabel('Eje z')
ax.set_title('Curva 3D')
plt.show()
```

### Gráficas de superficies

$f: \mathbb{R}^2\rightarrow\mathbb{R}$


```python
x = linspace(-3, 3)
y = linspace(-2, 2)
X, Y = meshgrid(x,y) #Numpy
Z = (X*Y**2)/(X**2+Y**2)
ax = plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z)
plt.show()
```
