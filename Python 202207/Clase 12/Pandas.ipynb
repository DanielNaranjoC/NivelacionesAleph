## Manipulando datos

```python
import pandas as pd
datos = pd.read_excel('../datos.xlsx', sheet_name=1, index_col=0)
datos
```

### Creaci√≥n de columnas, aplicar funciones

```python
def grupo(x):
    if x < -1:
        return 'A'
    elif x < 0:
        return 'B'
    elif x < 1:
        return 'C'
    else:
        return 'D'

datos['X1*'] = datos['X1'].apply(grupo)
datos
```

### Creaci√≥n de tablas de frecuencia

```python
print(datos['X1*'].value_counts())
print(datos['X1*'].value_counts() / len(datos['X1*']))
```

### Agrupaci√≥n de datos

```python
datos.groupby('X1*').sum()
```

```python
datos.groupby('X1*').mean()
```

```python
datos.groupby('X1*').var()
```

### Uni√≥n de bases de datos

```python
datos_2 = pd.read_excel('../datos.xlsx', sheet_name=2, index_col=0)
datos_2
```

```python
pd.concat([datos, datos_2])
```

```python
pd.concat([datos, datos_2], axis=1)
```

```python
datos.join(datos_2)
```
