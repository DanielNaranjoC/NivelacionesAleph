## Graficación

Tomemos en cuenta la curva paramétrica tal que: \begin{align*} x(t)&=\sin(t)\left(e^{\cos(t)}-2\cos(4t)-\sin ^{5}\left(\dfrac{t}{12}\right)\right)\\ y(t)&=\cos(t)\left(e^{\cos(t)}-2\cos(4t)-\sin ^{5}\left(\dfrac{t}{12}\right)\right) \end{align*}

Graficar esta curva con $0\leq t\leq 2\pi$ y con $0\leq t\leq 10\pi$ en una misma ventana.

```python
from numpy import *
from matplotlib.pyplot import *
```

```python
%matplotlib notebook
t1 = linspace(0,2*pi,150)
t2 = linspace(0,10*pi,1000)
x1 = sin(t1)*(exp(cos(t1))-2*cos(4*t1)-sin(t1/12)**5)
x2 = sin(t2)*(exp(cos(t2))-2*cos(4*t2)-sin(t2/12)**5)
y1 = cos(t1)*(exp(cos(t1))-2*cos(4*t1)-sin(t1/12)**5)
y2 = cos(t2)*(exp(cos(t2))-2*cos(4*t2)-sin(t2/12)**5)

subplot(1,2,1)
plot(x1,y1)
subplot(1,2,2)
plot(x2,y2)
show()
```
