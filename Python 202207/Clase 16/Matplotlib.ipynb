```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi
from scipy.stats import norm
```

## Otros comandos de graficaci√≥n

```python
x = np.linspace(-pi, pi, 500)
y1 = x**2 -2 * x + 1
y2 = np.tan(x)
y3 = np.cos(x)
plt.plot(x,y1,x,y2,x,y3)
plt.axis([-pi, pi, -pi, 2*pi])
plt.show()
```

```python
x = np.linspace(-pi, pi, 500)
y1 = x**2 -2 * x + 1
y2 = np.tan(x)
y3 = np.cos(x)
plt.plot(x,y1,x,y2,x,y3)
plt.axis([-pi, pi, -pi, 2*pi])

plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Funciones varias: $f_1,\ f_2,\ f_3$')
plt.show()
```

## Varios planos en una figura

```python
x = np.linspace(-pi, pi, 500)
y1 = x**2 -2 * x + 1
y2 = np.tan(x)
y3 = np.cos(x)

plt.subplot(1, 3, 1)
plt.plot(x, y1)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('$x^2-2x+1$')
plt.axis([-pi, pi, min(y1), max(y1)])

plt.subplot(1, 3, 2)
plt.plot(x, y2)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title(r'$\tan{(x)}$')
plt.axis([-pi, pi, -2*pi, 2*pi])

plt.subplot(1, 3, 3)
plt.plot(x, y3)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('$\cos{(x)}$')
plt.axis([-pi, pi, -1, 1])

plt.show()
```

## Otros tipos de gr√°ficos
### Dispersi√≥n

```python
x = -5 + 10*np.random.random(20)
y = 3 + x/2 + 2*np.random.random(20)
plt.scatter(x, y)
plt.show()
```

### Erorres

```python
x = np.arange(10, 24, 2)
y = np.array([950, 650, 210, 870, 1009, 755, 899])
errores = [abs(i) for i in 30*np.random.randn(7)]
plt.errorbar(x,y,yerr=errores)
plt.show()
```

### Histograma

```python
x = np.random.randn(1000)
plt.hist(x,bins=20)
plt.show()
```

```python
x = np.random.randn(10000)
x1 = np.linspace(min(x), max(x))
d = norm.pdf(x1)
n, bins, _ = plt.hist(x, bins=25, density=True)
plt.plot(x1, d)
aux_b = [(bins[i] + bins[i+1])/2 for i in range(len(n))]
plt.plot(aux_b, n)
plt.show()
```

### Boxplot

```python
x = np.random.randn(1000)
plt.boxplot(x)
plt.show()
```

### Barras

```python
x = np.arange(2000, 2020, 2)
y = 15 + np.random.randn(10).cumsum()
plt.bar(x, y)
plt.show()
```

```python
x = np.arange(2000, 2020, 2)
y = 15 + np.random.randn(10).cumsum()
plt.barh(x, y)
plt.show()
```

```python
x = np.arange(2000, 2020, 2)
y = 15 + np.random.randn(10).cumsum()
plt.step(x, y)
plt.show()
```

```python
x = np.arange(2000, 2020, 2)
y = 15 + np.random.randn(10).cumsum()
plt.stem(x, y)
plt.show()
```
