## Comparar
Comparar el resultado de mi función raíz cuadrada con el resultado de python

```python
from math import sqrt
def raiz(a, tolerancia=1e-6):
    x = a
    error = abs(x**2 - a)
    while error > tolerancia:
        x = x - (x**2-a)/(2*a)
        error = abs(x**2 - a)
    return x

for i in range(20):
    aux_1 = raiz(i, 1e-10)
    aux_2 = sqrt(i)
    print('Mi raíz:', aux_1, '\tRaíz python:', aux_2, '\tDiferencia:', abs(aux_1-aux_2))
```

## Funciones y módulos
Crear una función que retorne una lista de números aleatorios con distribución normal

```python
from random import gauss
def normal(n, mu=0, sigma=1):
    list_n = []
    for i in range(n):
        list_n.append(gauss(mu, sigma))
    return list_n
for i in range(7):
    print(normal(i))
```

Crear una función que retorne una lista de números aleatorios con distribución exponencial

```python
from random import expovariate
def exp(n, lambda_e=1):
    list_n = []
    for i in range(n):
        list_n.append(expovariate(lambda_e))
    return list_n
for i in range(7):
    print(exp(i))
```

Crear una función que reciba dos números n y m, de manera que retorne una lista de tamaño n con números primos aletorios menores o iguales a m

```python
from random import randint
def EsPrimo(n):
    if n < 2:
        return False
    primo = True
    for i in range(2, n):
        if n % i == 0:
            primo = False
            break
    return primo
def primos_rand(n, m):
    list_p = []
    if m < 2:
        return list_p
    while len(list_p) < n:
        aux = randint(2, m)
        if EsPrimo(aux):
            list_p.append(aux)
    return list_p
for i in range(7):
    print(primos_rand(i, 20))
```
